```python
import pandas as pd
import networkx as nx
```

```python
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}
G = nx.Graph()
G.add_nodes_from(graph)
for g_item in graph:
    edges = list(map(lambda x: (g_item, x), graph[g_item]))
    G.add_edges_from(edges)

nx.draw(G, with_labels=True)
```

```python
num_nodes = G.number_of_nodes()
num_edges = G.number_of_edges()
is_connected = nx.is_connected(G)

print(f"Кількість нод: {num_nodes}\nКількість ребер: {num_edges}\nЗв'язний граф: {is_connected}")
degree_of_nodes = dict(G.degree())

print(f"Ступінь вершин: {degree_of_nodes}")
```
